```python
#Analysing association between the categorical variables
from scipy.stats import chi2_contingency
def cramers_v(x, y):
    contin_tab = pd.crosstab(x, y)
    chi2, p, dof, exp = chi2_contingency(contin_tab)
    n = contin_tab.sum().sum()
    k = min(contin_tab.shape)
    dof = k-1


    return np.sqrt(chi2/(n*dof))
```

```python
#Creating the Association Matrix
categorical_columns = df.select_dtypes(include='object').columns
num_categorical = len(categorical_columns)
association_matrix = np.zeros((num_categorical, num_categorical))

for i in range(num_categorical):
    for j in range(num_categorical):
        if i != j:
            v = cramers_v(df
          [categorical_columns[i]], df
          [categorical_columns[j]])
            association_matrix[i, j] = v

association_df = pd.DataFrame(association_matrix, index=categorical_columns, columns=categorical_columns)
print("Association Matrix of Cramer's V:")
print(association_df)
```

1. No significant association between any categorical variables.
2. But there some moderate association of Gender column and discount applied column.
